```python
import pandas as pd

# URL del dataset
url = "supermarket_sales.csv"

# Cargar dataset
df = pd.read_csv(url)

# Mostrar primeras filas
df.head()
```

```python
# Dimensiones del dataset
print("Filas y columnas:", df.shape)

# Tipos de datos
df.info()
```

```python
# Convertir columnas de fecha y hora
df["Date"] = pd.to_datetime(df["Date"])
df["Time"] = pd.to_datetime(df["Time"]).dt.time

# Crear columna Año-Mes (para análisis temporal)
df["YearMonth"] = df["Date"].dt.to_period("M")
```

```python
# Asegurar conversión correcta de Time
df["Time"] = pd.to_datetime(df["Time"], format="%H:%M:%S").dt.time

# -----------------------------
# CAMPOS BÁSICOS
# -----------------------------
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Hour"] = pd.to_datetime(df["Time"], format="%H:%M:%S").dt.hour

# -----------------------------
# CAMPOS AVANZADOS
# -----------------------------

# Día de la semana (0=lunes, 6=domingo)
df["Weekday"] = df["Date"].dt.weekday

# Nombre del día
df["Weekday_Name"] = df["Date"].dt.day_name()

# Nombre del mes
df["Month_Name"] = df["Date"].dt.month_name()

# Trimestre (1, 2, 3, 4)
df["Quarter"] = df["Date"].dt.quarter

# Año-Mes en formato “YYYY-MM”
df["YearMonth"] = df["Date"].dt.to_period("M")

# Franja horaria
def franja_horaria(h):
    if 6 <= h < 12:
        return "Mañana"
    elif 12 <= h < 18:
        return "Tarde"
    elif 18 <= h < 24:
        return "Noche"
    else:
        return "Madrugada"

df["Time_Block"] = df["Hour"].apply(franja_horaria)
```

```python
# Top 3 product lines por facturación total
top3_product_lines = (
    df.groupby("Product line")["Total"]
      .sum()
      .sort_values(ascending=False)
      .head(3)
)

top3_product_lines
```

```python
# Facturación por ciudad
facturacion_ciudad = (
    df.groupby("City")["Total"]
      .sum()
      .sort_values(ascending=False)
)

facturacion_ciudad
```

```python
# Ticket promedio = promedio del total de cada invoice
ticket_promedio = df["Total"].mean()
ticket_promedio
```

```python
# Curva mensual de ventas
curva_mensual = (
    df.groupby("YearMonth")["Total"]
      .sum()
      .sort_values()
)

curva_mensual
```
